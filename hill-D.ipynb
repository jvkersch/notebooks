```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
q = np.linspace(0, 15, 50)
```

```python
p1 = (1/10)**q
p2 = (9/10)**q
D = (p1+p2)**(1/(1-q))
```

```python
plt.plot(q, p1, label=r"$(1/10)^q$", linestyle="dashed", color="gray")
plt.plot(q, p2, label=r"$(9/10)^q$", linestyle="dotted", color="gray")
plt.plot(q, D, label="Hill D(q)")
plt.xlabel("q")

plt.legend()
```
