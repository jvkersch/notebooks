```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
answers = (pd.read_csv("answers-float.csv")
    .rename(columns={"a*b": "chatgpt"}))
answers["exact"] = answers.a * answers.b
answers["abs_diff"] = np.abs(answers["exact"] - answers["chatgpt"])
answers["rel_diff"] = answers.abs_diff / answers["exact"]
```

```python
answers
```

```python
sns.stripplot(
    data=answers,
    x="b",
    y="abs_diff",
    jitter=False,
)
plt.yscale("log")
```

```python
sns.stripplot(
    data=answers,
    x="b",
    y="rel_diff",
    jitter=False,
)
plt.yscale("log")
```
