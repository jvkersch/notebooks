```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

## Dirichlet distributions

### The role of the concentration parameter: location

```python
from scipy.stats import dirichlet
```

```python
def make_data(alpha):
    x = np.linspace(1e-5, 1, 50)
    y = np.linspace(1e-5, 1, 50)
    X, Y = np.meshgrid(x, y)
    Z = 1 - X - Y
    simplex_mask = Z >= 0

    Xs = X[simplex_mask]
    Ys = Y[simplex_mask]
    Zs = Z[simplex_mask]

    values = np.full_like(X, np.nan)
    values[simplex_mask] = dirichlet.pdf(
        np.column_stack((Xs, Ys, Zs)).T, alpha
    )
    return X, Y, values
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

alpha = (7, 3, 3)
X, Y, v = make_data(alpha)
axes[0,0].contourf(X, Y, v, 20, cmap='RdGy_r')
axes[0,0].set_title(f"alpha = {alpha}")

alpha = (3, 7, 3)
X, Y, v = make_data(alpha)
axes[0,1].contourf(X, Y, v, 20, cmap='RdGy_r')
axes[0,1].set_title(f"alpha = {alpha}")

alpha = (3, 3, 7)
X, Y, v = make_data(alpha)
axes[1,0].contourf(X, Y, v, 20, cmap='RdGy_r')
axes[1,0].set_title(f"alpha = {alpha}")

alpha = (3, 3, 3)
X, Y, v = make_data(alpha)
axes[1,1].contourf(X, Y, v, 20, cmap='RdGy_r')
axes[1,1].set_title(f"alpha = {alpha}")

pass
```
